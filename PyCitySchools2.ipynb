```python
#CSV file paths
schools_csv = "/users/rebeccatast/desktop/Pandas-Challenge/Resources/PyCitySchools.csv"
students_csv = "/users/rebeccatast/desktop/Pandas-Challenge/Resources/PyCityStudents.csv"
```

```python
#Set dataframes
schools_df = pd.read_csv(schools_csv)
students_df = pd.read_csv(students_csv)
```

```python
#Merge datasets and print
complete_data_df = pd.merge(schools_df, students_df,how = 'left', on=["school_name","school_name"])
print (complete_data_df)
complete_data_df.head()
```

```python
#Calculate total students, total schools, total budget
total_schools = len(complete_data_df["School ID"].unique())
print(total_schools)
total_students = complete_data_df["Student ID"].count()
print(total_students)
total_budget = sum(complete_data_df ["budget"].unique())
print(total_budget)
```

```python
#Calculate Average Math and Reading Scores
avg_math = complete_data_df["math_score"].mean()
print(avg_math)
avg_reading = complete_data_df ["reading_score"].mean()
print(avg_reading)
```

```python
#Calculate percentage of students passing math
passing_math = complete_data_df.loc[students_df["math_score"] >= 70] ["math_score"].count()
print(passing_math)
percent_pass_math = passing_math/(total_students) *100
print(percent_pass_math)
```

```python
#Calculate percentage of students passing reading
passing_reading = complete_data_df.loc[students_df["reading_score"] >= 70] ["reading_score"].count()
print(passing_reading)
percent_pass_reading = passing_reading/(total_students) *100
print(percent_pass_reading)
```

```python
#Calculate total that passed math and reading
total_pass_rate = students_df[(students_df["math_score"] >= 70) & (students_df["reading_score"] >= 70)]["student_name"].count()/float(total_students)*100
print(total_pass_rate)
```

```python
#District Summary Dataframe
district_summary = pd.DataFrame({"Total Schools": [total_schools],
                           "Total Students": [total_students],
                           "Total Budget": [total_budget],
                           "Average Math Score": [avg_math],
                           "Average Reading Score": [avg_reading],
                           "% Students Passing Math":[percent_pass_math],
                           "% Students Passing Reading": [percent_pass_reading],
                           "% Students Passing Overall": [total_pass_rate]})



district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Students Passing Math"] = district_summary["% Students Passing Math"].map("{:.2f}%".format)
district_summary["% Students Passing Reading"] = district_summary["% Students Passing Reading"].map("{:.2f}%".format)
district_summary["% Students Passing Overall"] = district_summary["% Students Passing Overall"].map("{:.2f}%".format)

district_summary
```
